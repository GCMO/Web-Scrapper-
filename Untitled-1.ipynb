### THIS IS A PROGRAM TO SCRAPE COMPANY NAME & STOCK TICKERS FROM A WEBSITE


```python
# IMPORT LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import bs4 as bs
import pickle
import requests

```

```python
# Create 3 lists: for the Company Name, the Company Ticker and the Stock Price
CName = []
CTicker = []
CPrice = [] # we create this list even though the data is non available with these links
```

```python
# scrape the data from advfn.com - 
# note this link only collect all the Company Names starting with "A" you'll all to scrape all the other letters with a loop
url='https://www.advfn.com/nyse/newyorkstockexchange.asp?companies=A'
page=requests.get(url)
soup=bs.BeautifulSoup(page.text,'html.parser')
odd_rows=soup.find_all('tr', attrs={'class':'ts0'})
even_rows=soup.find_all('tr', attrs={'class':'ts1'})
```

```python
# clean the data and append to each respective list
for i in odd_rows:
    row = i.find_all('td')
    CName.append(row[0].text.strip())
    CTicker.append(row[1].text.strip())
    CPrice.append(row[2].text.strip())
    # print (CName, CTicker, CTicker) (this shows them all in the console)

for i in even_rows:
    row = i.find_all('td')
    CName.append(row[0].text.strip())
    CTicker.append(row[1].text.strip())
    CPrice.append(row[2].text.strip())
```

```python
# create a dataframe with the scraped data
df = pd.DataFrame({'Company Name':CName, 'Company Ticker':CTicker, 'Stock Price':CPrice})
df

```

```python
# Create a function to scrape the data from advfn.com
def scrape_data(url):
    # Create a variable to store the data
    data = []
    # Get the data from the url
    sauce = requests.get(url).text
    # Create a soup object
    soup = bs.BeautifulSoup(sauce, 'lxml')
    # Get the table
    table = soup.find('table', {'class': 'W(100%) M(0)'})
    # Get the rows
    rows = table.find_all('tr')
    # Loop through the rows
    for row in rows:
        # Get the columns
        cols = row.find_all('td')
        # Loop through the columns
        for col in cols:
            # Get the text
            text = col.text.strip()
            # Append the text to the data
            data.append(text)
    # Return the data
    return data

scrape_data("https://advfn.com/nyse/newyorkstockexchange.asp?companies=A")
```

```python
#create a function to scrape the data from alpaca.com and return the data
def scrape_data_alpaca(url):
    # Create a list to store the data
    data = []
    # Get the data from the url
    sauce = requests.get(url).text
    # Create a soup object
    soup = bs.BeautifulSoup(sauce, 'lxml')
    # Get the table
    table = soup.find('table', {'class': 'W(100%) M(0)'})
    # Get the rows
    rows = table.find_all('tr')
    # Loop through the rows
    for row in rows:
        # Get the columns
        cols = row.find_all('td')
        # Loop through the columns
        for col in cols:
            # Get the text
            text = col.text.strip()
            # Append the text to the data
            data.append(text)
    # Return the data
    return data

scrape_data_alpaca("https://www.alpaca.markets/stocks/A")
```
